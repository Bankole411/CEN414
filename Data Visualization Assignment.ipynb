# Enrolment-in-Secondary-Education

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import statistics
```

```python
af_countries = [    'Algeria',
                    'Angola',
                    'Benin',
                    'Botswana',
                    'Burkina Faso',
                    'Burundi',
                    'Cameroon',
                    'Cape Verde',
                    'Central African Republic',
                    'Chad',
                    'Comoros',
                    "Côte d'Ivoire",
                    'Democratic Republic of the Congo',
                    'Djibouti',
                    'Egypt',
                    'Equatorial Guinea',
                    'Eritrea',
                    'Eswatini',
                    'Ethiopia',
                    'Gabon',
                    'Gambia',
                    'Ghana',
                    'Guinea',
                    'Guinea-Bissau',
                    'Kenya',
                    'Lesotho',
                    'Liberia',
                    'Libyan Arab Jamahiriya',
                    'Madagascar',
                    'Malawi',
                    'Mali',
                    'Mauritania',
                    'Mauritius',
                    'Morocco',
                    'Mozambique',
                    'Namibia',
                    'Niger',
                    'Nigeria',
                    'Rwanda',
                    'Sao Tome and Principe',
                    'Senegal',
                    'Seychelles',
                    'Sierra Leone',
                    'Somalia',
                    'South Africa',
                    'South Sudan',
                    'Sudan',
                    'Tanzania',
                    'Togo',
                    'Tunisia',
                    'Uganda'
                    'Zambia',
                    'Zimbabwe'
        ]

df   = pd.read_csv('data.csv')
df = df[ (df['Reference Area'].isin(af_countries) )]
df = df.sort_values('Observation Value')
df = df.set_index('Reference Area')
df
```

```python
for i in range(5):
    l = 2000+i*3
    h = l + 3
    fig, axes  = plt.subplots(1,3, figsize=(10,10))
    plt.subplots_adjust(right=2)
    
    for idx, year in enumerate(range(l,h)):
        df[(df['Time Period']== year) & (df['Sex'] == 'All genders')]['Observation Value'].plot.barh(ax=axes[idx])
        axes[idx].set_title(year)
        axes[idx].set_ylabel('')  
        
```

## Yearly Observations

```python
mean_vals = []
for year in range(2000,2016):
    mean_vals.append(df[df['Time Period'] == year]['Observation Value'].mean())
    
plt.plot(range(2000,2016), mean_vals)
plt.title('Mean Enrolment-in-Secondary-Education of African Countries 2000-2015')
plt.xlabel('Year')
plt.ylabel('Percentage')
```

```python
median_vals = []
for year in range(2000,2016):
    median_vals.append(df[df['Time Period'] == year]['Observation Value'].median())
    
plt.plot(range(2000,2016), median_vals)
plt.title('Median Enrolment-in-Secondary-Education of African Countries 2000-2016')
plt.xlabel('Year')
plt.ylabel('Percentage')
```

```python
var_vals = []
for year in range(2000,2016):
    var_vals.append(df[df['Time Period'] == year]['Observation Value'].var())
    
plt.plot(range(2000,2016), var_vals)
plt.title('Variance of Enrolment-in-Secondary-Education of African Countries 2000-2016')
plt.xlabel('Year')
plt.ylabel('Percentage')
```

```python
stdev_vals = []
for year in range(2000,2016):
    stdev_vals.append(df[df['Time Period'] == year]['Observation Value'].var()**0.5)
    
plt.plot(range(2000,2016), stdev_vals)
plt.title('Standard Deviation of Enrolment-in-Secondary-Education of African Countries 2000-2016')
plt.xlabel('Year')
plt.ylabel('Percentage')
```

```python
range_vals = []
for year in range(2000,2016):
    range_vals.append(df[df['Time Period'] == year]['Observation Value'].max() - df[df['Time Period'] == year]['Observation Value'].min())

    
plt.title('Range of Enrolment-in-Secondary-Education Values for African Countries 2000-2016')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.plot(range(2000,2016), range_vals)
```

```python
iqr_vals = []
def get_iqr(sample):
    q3,q1 = np.percentile(sample, [75 ,25])
    return q3 - q1

for year in range(2000,2016):
    iqr_vals.append(get_iqr(df[df['Time Period'] == year]['Observation Value']))
    
    
plt.title('Inter Quartile Range Values of Enrolment-in-Secondary-Education of African Countries 2000-2016')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.plot(range(2000,2016), iqr_vals)
```

```python
siqr_vals = []

for year in range(2000,2016):
    siqr_vals.append(get_iqr(df[df['Time Period'] == year]['Observation Value'])*0.5)
    
    
plt.title('Semi Inter-Quartile Range Values of Enrolment-in-Secondary-Education of African Countries 2000-2016')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.plot(range(2000,2016), siqr_vals)
```
